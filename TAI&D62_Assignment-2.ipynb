# Imported Important Libraries

```python
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from warnings import filterwarnings
filterwarnings('ignore')
```

# Generating dataset

```python
Math_score = [random.randint(60, 80) for _ in range(400)]
missing_pct = 0.2  # 20% missing values

# Calculate the number of missing values
n_missing = int(len(Math_score) * missing_pct)

# Generate a list of indices to replace with missing values
missing_indices = random.sample(range(len(Math_score)), n_missing)

# Replace the values at the missing indices with None
for i in missing_indices:
    Math_score[i] = None
```

```python
Reading_score = [random.randint(75, 95) for _ in range(400)]
missing_pct = 0.1  # 10% missing values

# Calculate the number of missing values
n_missing = int(len(Reading_score) * missing_pct)

# Generate a list of indices to replace with missing values
missing_indices = random.sample(range(len(Reading_score)), n_missing)

# Replace the values at the missing indices with None
for i in missing_indices:
    Reading_score[i] = None
```

# Detecting outliers

```python
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(dataFrame['Math_score'])

plt.subplot(2,2,2)
sns.boxplot(dataFrame['Math_score'])
```

# Finding the IQR(Inter Quartile Range)

```python
percentile25 = dataFrame['Math_score'].quantile(0.25)
percentile75 = dataFrame['Math_score'].quantile(0.75)
```

```python
percentile25
```

```python
percentile75
```

```python
iqr = percentile75 - percentile25
iqr
```

```python
upper_limit = percentile75 + 1.5 * iqr
lower_limit = percentile25 - 1.5 * iqr
```

```python
print("Upper limit",upper_limit)
print("Lower limit",lower_limit)
```

# Finding outliers

```python
dataFrame[dataFrame['Math_score'] > upper_limit]
```

```python
dataFrame[dataFrame['Math_score'] < lower_limit]
```

# Trimming


```python
new_dataFrame = dataFrame[dataFrame['Math_score'] > lower_limit]
new_dataFrame.shape
```

```python
# Comparing

plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(dataFrame['Math_score'])

plt.subplot(2,2,2)
sns.boxplot(dataFrame['Math_score'])

plt.subplot(2,2,3)
sns.distplot(new_dataFrame['Math_score'])

plt.subplot(2,2,4)
sns.boxplot(new_dataFrame['Math_score'])

plt.show()
```

# Capping

```python
new_df_cap = dataFrame.copy()

new_df_cap['Math_score'] = np.where(
    new_df_cap['Math_score'] > upper_limit,
    upper_limit,
    np.where(
        new_df_cap['Math_score'] < lower_limit,
        lower_limit,
        new_df_cap['Math_score']
    )
)
```

```python
new_df_cap.shape
```

```python
# Comparing

plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(dataFrame['Math_score'])

plt.subplot(2,2,2)
sns.boxplot(dataFrame['Math_score'])

plt.subplot(2,2,3)
sns.distplot(new_df_cap['Math_score'])

plt.subplot(2,2,4)
sns.boxplot(new_df_cap['Math_score'])

plt.show()
```
